# CycleGAN

```python
import torch
from data.data_loader import TestDataSet
from model.cycle_gan import CycleGAN
from torch.utils.data import DataLoader
from torch.autograd import Variable
from torch import Tensor
import matplotlib.pyplot as plt
%matplotlib notebook
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import numpy as np
```

### define parameters

```python
models_prefix = "./model/saved_models/test_"
data_dir = "./data/"
models_dir = "./model/saved_models/test_"

image_size = [256, 256]
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```

### define model

```python
cycle_gan = CycleGAN(device, models_dir, train=False)
```

### load model

```python
dataset = TestDataSet(data_root=data_dir, image_size=(image_size[0], image_size[1]))
dataloader = DataLoader(dataset, batch_size=1, shuffle=False)
```

```python
def myimshow(image, ax=plt):
    image = image.to('cpu').numpy()
    image = np.moveaxis(image, [0, 1, 2], [2, 0, 1]) 
    image = (image + 1) / 2
    image[image < 0] = 0
    image[image > 1] = 1
    h = ax.imshow(image)
    ax.axis('off')
    return h
```

```python
for i,data in enumerate(dataloader):
    
    x = Variable(data.type(Tensor))

    cycle_gan.set_input(x, x)
    cycle_gan.test()
    gen_image = cycle_gan.get_generated_image()
#     gen_image = np.moveaxis(gen_image, 0,-1)
#     print(gen_image)
    fig = plt.figure() 
    myimshow(gen_image[0])
#     
#     plt.imshow(gen_image) 
```
